```python
netflix['computer_date'] = pd.to_datetime(netflix['date_added'], infer_datetime_format=True, errors='coerce')
netflix
```

```python
from datetime import datetime as dt
```

```python
netflix["computer_date"].fillna(dt(2020, 1, 1), inplace=True)
netflix
```

```python
netflix["unix_time"] = netflix["computer_date"].astype(int)
```

```python
netflix
```

```python
netflix[["Date Part 1", "Date Part 2"]] = netflix["date_added"].str.split(", ", expand=True)
netflix
```

```python
netflix.rename(columns={"Date Part 2": "Year"}, inplace=True)
```

```python
netflix
```

```python
netflix["cast"].fillna(value="no cast", inplace=True)
netflix
```

```python
netflix["genre_count"] = netflix["listed_in"].map(lambda x: len(x.split(",")))
netflix
```

```python
netflix_movies_by_counrties = netflix.groupby("country")["show_id"].count().reset_index()
netflix_movies_by_counrties.sort_values(by="show_id", ascending=False)
```
